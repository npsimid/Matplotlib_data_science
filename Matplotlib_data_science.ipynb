### 3.3 Grafice -  bare de nivel - metoda `bar()` sau `barh()`

```python
# Crearea unor date in cadrul unui dictionar
prezenta = {"MMRT": 10,
            "SISRC": 12,
            "SCE": 8}
prezenta
```

```python
# constructia graficului prezentei sub forma de bare verticale
fig, ax = plt.subplots()
ax.bar(prezenta.keys(), prezenta.values())
ax.set(title="Prezenta studentilor la lectie",
       xlabel="Grupa",
       ylabel="Numarul de studenti");
```

```python
# constructia graficului prezentei sub forma de bare verticale
fig, ax = plt.subplots()
ax.barh(list(prezenta.keys()), list(prezenta.values()))
ax.set(title="Prezenta studentilor la lectie",
       ylabel="Grupa",
       xlabel="Numarul de studenti");
```

### 3.4 Crearea unei figuri cu subgrafice

```python
# Optiunea 1: Crearea subgraficelor multiple

# 1 Setarea figurii si sub-graficelor
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2, 
                                             figsize=(10, 7))
# Afisarea datelor pe fiecare axa
ax1.plot(x, np.sin(x));
ax2.scatter(x, np.exp(x));
ax3.bar(prezenta.keys(), prezenta.values());
ax4.hist(y, bins=20);
```

```python
# Optiunea 2: Crearea subgraficelor multiple

# 1 Setarea figurii si sub-graficelor
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Afisarea datelor pe fiecare axa
ax[0,0].plot(x, np.sin(x));
ax[0,1].scatter(x, np.exp(x));
ax[1,0].bar(prezenta.keys(), prezenta.values());
ax[1,1].hist(y, bins=30);
```

## 4. Afisarea datelor din Pandas

### 4.1. Importul si pregatirea datelor

```python
# importul setului de date Vanzari_masini
vanzari_masini = pd.read_csv("Vanzari_masini.csv")
vanzari_masini
```

```python
#Stergerea simbolurilor $, , ;i . din valorile coloanei Pret 
vanzari_masini["Pret"] = vanzari_masini["Pret"].str.replace('[\$\,\.]', '')
vanzari_masini
```

```python
#Stergerea a 2 zerouri de la urma pentru valorile coloanei Pret
vanzari_masini["Pret"] = vanzari_masini["Pret"].str[:-3]
vanzari_masini
```

```python
# determinarea tipului de date in coloane
vanzari_masini.dtypes
```

```python
#Transformarea datelor coloanei Pret in tip intreg
vanzari_masini["Pret"] = vanzari_masini["Pret"].astype(int)
vanzari_masini
```

```python
# determinarea tipului de date in coloane
vanzari_masini.dtypes
```

```python
# Adaugarea unei coloane in care se va specifica data vanzarii
vanzari_masini["Data vanzare"] = pd.date_range("1/1/2021", periods=len(vanzari_masini))
vanzari_masini
```

```python
# Adaugarea unei coloane in care se va specifica suma de bani acumulata
vanzari_masini["Total bani"] = vanzari_masini["Pret"].cumsum()
vanzari_masini
```

#### 4.2.1. Afisarea cu linie continua

```python
# afisarea grafica a dependentei valorilor coloanei Total bani de coloana Data vanzare
vanzari_masini.plot(x='Data vanzare', y='Total bani');
```

#### 4.2.2. Afisarea cu linie punctata

```python
# afisarea grafica a dependentei valorilor coloanei Pret de coloana Kilometraj (KM)
vanzari_masini.plot(x="Kilometraj (KM)", y="Pret", kind='scatter');
```

#### 4.2.3. Afisarea cu bare de nivel

```python
# afisarea grafica a dependentei valorilor coloanei Kilometraj (KM) de coloana Producator
vanzari_masini.plot(x='Producator', y='Kilometraj (KM)', kind='bar');
```

#### 4.2.4. Afisarea histogramelor

```python
# optiunea 1: afisarea histogramei coloanei Kilometraj (KM)
vanzari_masini["Kilometraj (KM)"].plot(kind="hist");
```

```python
# optiunea 2: afisarea histogramei coloanei Kilometraj (KM)
vanzari_masini["Kilometraj (KM)"].plot.hist();
```

```python
# afisarea histogramei coloanei Kilometraj (KM) cu un numar mai mare de intervale 
vanzari_masini["Kilometraj (KM)"].plot.hist(bins=20);
```

### 4.3. Afisarea grafica complexa a datelor

```python
# importul setului de date heart_disease
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
# afisarea depedentei coloanei chol de colana age, iar culoarea punctelor 
# va depinde de valoarea colaonei target
heart_disease.plot(kind='scatter', 
             x='age', 
             y='chol', 
             c='target', 
             figsize=(10, 6));
```

```python
# Imbunatatirea graficului de mai sus

# 1 Setarea figurii si sub-graficelor
fig, ax = plt.subplots(figsize=(10, 6))

# 2 afisarea datelor
scatter=ax.scatter(x=heart_disease["age"], 
                   y=heart_disease["chol"], 
                   c=heart_disease["target"])

# 3 Personalizarea graficului
ax.set(title="Boala de inima in functie de nivelul de colesterol",
       xlabel="Varsta",
       ylabel="Nivelul de colesterol");

# 4 Adaugarea legendei
ax.legend(*scatter.legend_elements(), title="Prezenta bolii");
```

```python
# Imbunatatirea graficului de mai sus

# 1 Setarea figurii si sub-graficelor
fig, ax = plt.subplots(figsize=(10, 6))

# 2 afisarea datelor
scatter=ax.scatter(x=heart_disease["age"], 
                   y=heart_disease["chol"], 
                   c=heart_disease["target"])

# 3 Personalizarea graficului
ax.set(title="Boala de inima in functie de nivelul de colesterol",
       xlabel="Varsta",
       ylabel="Nivelul de colesterol");

# 4 Adaugarea legendei
ax.legend(*scatter.legend_elements(), title="Prezenta bolii");

# 5  Adaugarea linie cu valoarea medie a nivelului de colesterol
ax.axhline(heart_disease["chol"].mean(),
           linestyle="--");
```

```python
# afisarea depedentei coloanlor chol si thalach de colana age, iar culoarea punctelor 
# va depinde de valoarea coloanei target

# 1 Setarea figurii si sub-graficelor
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1, 
                               sharex=True, 
                               figsize=(10, 8))

# 2 afisarea datelor pe axa 0
scatter=ax0.scatter(x=heart_disease["age"], 
                    y=heart_disease["chol"], 
                    c=heart_disease["target"])

# 3 Personalizarea graficului pe axa 0
ax0.set(title="Boala de inima in functie de nivelul de colesterol",
        ylabel="Nivelul de colesterol")

# 4 Adaugarea legendei pe axa 0
ax0.legend(*scatter.legend_elements(), title="Prezenta bolii")

# 5  Adaugarea linie cu valoarea medie a nivelului de colesterol pe axa 0
ax0.axhline(y=heart_disease["chol"].mean(),  
            linestyle='--')

# 6 afisarea datelor pe axa 1
scatter=ax1.scatter(x=heart_disease["age"], 
                    y=heart_disease["thalach"], 
                    c=heart_disease["target"])

# 7 Personalizarea graficului pe axa 1
ax1.set(title="Boala de inima in functie numarul maxim a batailor de inima",
        xlabel="Varsta",
        ylabel="Numarul maxim batai")

# 8 Adaugarea legendei pe axa 1
ax1.legend(*scatter.legend_elements(), title="Prezenta bolii")

# 9  Adaugarea linie cu valoarea medie a numarului maxi a batailor de inima pe axa 1
ax1.axhline(y=heart_disease["thalach"].mean(), 
            linestyle='--')

# 10 setarea titlului figurii
fig.suptitle('Analiza bolilor de inima', fontsize=16, fontweight='bold');
```

## 6. Salvarea graficelor

```python
# afisarea graficului ce va fi salvat

# 1 Setarea figurii si sub-graficelor
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1, 
                               sharex=True, 
                               figsize=(10, 8))

# 2 afisarea datelor pe axa 0
scatter=ax0.scatter(x=heart_disease["age"], 
                    y=heart_disease["chol"], 
                    c=heart_disease["target"],
                    cmap="winter")

# 3 Personalizarea graficului pe axa 0
ax0.set(title="Boala de inima in functie de nivelul de colesterol",
        ylabel="Nivelul de colesterol")

# 4 Adaugarea legendei pe axa 0
ax0.legend(*scatter.legend_elements(), title="Prezenta bolii")

# 5  Adaugarea linie cu valoarea medie a nivelului de colesterol pe axa 0
ax0.axhline(y=heart_disease["chol"].mean(),  
            linestyle='--')

# 6 afisarea datelor pe axa 1
scatter=ax1.scatter(x=heart_disease["age"], 
                    y=heart_disease["thalach"], 
                    c=heart_disease["target"],
                    cmap="spring")

# 7 Personalizarea graficului pe axa 1
ax1.set(title="Boala de inima in functie numarul maxim a batailor de inima",
        xlabel="Varsta",
        ylabel="Numarul maxim batai")

# 8 Adaugarea legendei pe axa 1
ax1.legend(*scatter.legend_elements(), title="Prezenta bolii")

# 9  Adaugarea linie cu valoarea medie a numarului maxi a batailor de inima pe axa 1
ax1.axhline(y=heart_disease["thalach"].mean(), 
            linestyle='--')

# 10 setarea titlului figurii
fig.suptitle('Analiza bolilor de inima', fontsize=16, fontweight='bold');
```
